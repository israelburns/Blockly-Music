```python
import re
import numpy as np
import pandas as pd
import json
```

```python
trace_filename = '../../trace.log'
```

```python
records = []
with open(trace_filename) as file:
    for line in file:
        m = re.match('.*timestampremove.*({.*e2e_millis.*})', line)
        if m:
            records += [json.loads(m.groups()[0])]
```

```python
df = pd.DataFrame(records)
```

```python
df.describe()
```

```python
df[['timestamp_millis']].plot(style='-', figsize=(20,8));
```
